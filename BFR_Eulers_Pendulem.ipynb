In class Euler Pendulum 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
theta = 11 * np.pi / 180  # convert to radians
omega = 0.0     # angular velocity
g = 9.81        # meters/second^2
l = 1.0         # meter
dt = 0.01       # time step in s (seconds)
t_max = 10      # seconds

# Lists to store history
t_list = []
theta_list = []
omega_list = []

# Initial conditions
t = 0
t_list.append(t)
theta_list.append(theta)
omega_list.append(omega)

# Simulation loop
while t < t_max:
    # Calculate angular acceleration (small angle approximation)
    alpha = (-g / l) * theta
    
    # Euler updates
    omega = omega + alpha * dt
    theta = theta + omega * dt
    t += dt
    
    # Store results
    t_list.append(t)
    theta_list.append(theta)
    omega_list.append(omega)

# Convert to arrays
t_array = np.array(t_list)
theta_array = np.array(theta_list)

# PLOT
plt.figure(figsize=(8,5))
plt.plot(t_array, theta_array, label='Simulation (Euler)')
plt.title("Pendulum motion: Simple Euler")
plt.xlabel("Time (s)")
plt.ylabel("Angle (rad)")
plt.axhline(0, color='green')
plt.grid()
plt.legend()
plt.show()
```
