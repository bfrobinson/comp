## Final Projct Qauntum Well

### Step One 
 Setting up the parmaters. Finding the work done on the system starts by taking the original single cell and finding the energy of it in ground state.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation 
import numpy as np

#constants 
hbar = 1.054e-34            #Js
m_e = 9.109e-31             #kg
eV = 1.602e-19              #J/eV

L = 1e-9  

#setting up the spatial grid learned from blank

N = 1000 
x = np.linspace(0, L, N) 

# Wave functions & energies  

def wf_infinite_well(x, n, L):
    return (2 / L) * np.sin(n * np.pi * x / L) 

def energy_infinite_well(n, L, m_e):       #energy eigenvalue for infinite well[0, L]
    return (n**2 * np.pi**2 * hbar**2) / (2 * m_e * L**2) 

# FINDING THE INTIAL IN THE GROUND STATE 

n_intial = 1
E_intial = energy_infinite_well(m_e, n_intial, L) 
wf_intial = wf_infinite_well(x, n_intial, L)

## Step one COMPLETE 

# STEP TWO Creating a barrier of the two wells of length L/2 

L_half = L / 2  
n_final = 1                 # it is in ground state for each half well 





def wf_infinite_well(x, n, L):
    return np.sqrt(2 / L) * np.sin(n * np.pi * x / L)

def energy_infinite_well(n, L, m_e):
    return (n**2 * np.pi**2 * hbar**2) / (2 * m_e * L**2)

n_initial = 1
E_initial = energy_infinite_well(n_initial, L, m_e)
wf_initial = wf_infinite_well(x, n_initial, L)

L_half = L / 2  
n_final = 1

E_left = energy_infinite_well(n_final, L_half, m_e)
E_right = energy_infinite_well(n_final, L_half, m_e)

E_final_avg = 0.5 * E_left + 0.5 * E_right
W = E_final_avg - E_initial  

##STEP TWO COMPLETE

print("Initial ground-state energy (full box): {:.3f} eV".format(E_initial / eV))
print("Ground-state energy (each half-box):   {:.3f} eV".format(E_left / eV))
print("Final average energy:                  {:.3f} eV".format(E_final_avg / eV))
print("Work done on system (final - initial): {:.3f} eV".format(W / eV))
E_right = energy_infinite_well(m_e , n_final , L_half)

#spilting it 50/50 
E_final_avg = 0.5 * E_left + 0.5 * E_right

#work work work 

W = E_final_avg - E_intial  

print("Initial ground-state energy (full box): {:.3f} eV".format(E_initial / eV))
print("Ground-state energy (each half-box):   {:.3f} eV".format(E_left / eV))
print("Final average energy:                  {:.3f} eV".format(E_final_avg / eV))
print("Work done on system (final - initial): {:.3f} eV".format(W / eV))
```

## Final wavefunctions

```python
#left

wf_left = np.zeros_like(x)
mask_left = (x >= 0) & (x <= L_half)
wf_left[mask_left] = wf_infinite_well(x[mask_left], n_final, L_half) 

#right side

wf_right = np.zeros_like(x)
mask_right = (x >= L_half) & (x <= L)
#  the right well runs from 0 to L/2
x_shift = x[mask_right] - L_half
wf_right[mask_right] = wf_infinite_well(x_shift, n_final, L_half) 

#assuring that the states are 50/50 
wf_final = 0.5 * np.abs(wf_left)**2 + 0.5 * np.abs(wf_right)**2

#PLOT 
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Initial probability density
axes[0].plot(x * 1e9, np.abs(wf_initial)**2, label=r'$|\psi_{\mathrm{initial}}|^2$')
axes[0].set_title("Initial: Full Box Ground State")
axes[0].set_xlabel("x (nm)")
axes[0].set_ylabel(r"Probability density")
axes[0].grid(True)

# Final probability density
axes[1].plot(x * 1e9, np.abs(wf_final)**2, label=r'$|\psi_{\mathrm{final}}|^2$', color='C1')
axes[1].axvline(L_half * 1e9, color='k', linestyle='--', label="Barrier")
axes[1].set_title("Final: Split Box (Barrier at Center)")
axes[1].set_xlabel("x (nm)")
axes[1].grid(True)

for ax in axes:
    ax.legend()

plt.tight_layout()
plt.show()





```

```python
# Creating plot of initial and final wavefunctions for a particle in a 1D infinite square well

import numpy as np
import matplotlib.pyplot as plt

# Set plot style


# Define constants
L = 1.0  # Width of the well
x = np.linspace(0, L, 1000)

# Define wavefunctions
psi1 = np.sqrt(2 / L) * np.sin(np.pi * x / L)       # n=1
psi2 = np.sqrt(2 / L) * np.sin(2 * np.pi * x / L)   # n=2

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x, psi1, label=r'$\psi_1(x) = \sqrt{2/L} \sin(\pi x / L)$', linewidth=2)
plt.plot(x, psi2, label=r'$\psi_2(x) = \sqrt{2/L} \sin(2\pi x / L)$', linewidth=2, linestyle='--')
plt.axvline(x=0.5, color='red', linestyle=':', label='Barrier at $x=0.5$')

# Add labels and title
plt.title('Initial and Final Wavefunctions in 1D Infinite Square Well', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel(r'$\psi(x)$', fontsize=12)
plt.legend()
plt.tight_layout()

# Save the plot

print("Plotted initial (n=1) and final (n=2) wavefunctions with barrier at x=0.5. Saved as wavefunctions_plot.png")
```
